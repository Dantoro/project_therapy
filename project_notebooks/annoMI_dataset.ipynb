```python
import pandas as pd
pd.set_option('display.max_colwidth',500)

# Path to AnnoMI data
AnnoMI_path = r'../datasets/custom_corpus/word2vec-sets/AnnoMI/AnnoMI-simple.csv'

# Read the AnnoMI dataset
AnnoMI_df = pd.read_csv(AnnoMI_path)

# Quick check

display(AnnoMI_df.head())

display(AnnoMI_df.info())

display(AnnoMI_df.describe())
```

```python
AnnoMI_df = AnnoMI_df[AnnoMI_df['mi_quality'] != "low"].reset_index(drop=True)
```

```python
display(AnnoMI_df.info())
```

```python
AnnoMI_df = AnnoMI_df[['topic', 'transcript_id', 'interlocutor', 'utterance_id', 'utterance_text']]
```

```python
AnnoMI_df.to_csv('annoMI_filtered.csv', index=False)
```

```python
# Find all conversations where the therapist is the first speaker
first_utterances = AnnoMI_df[AnnoMI_df['utterance_id'] == 0]
therapist_first_ids = first_utterances[first_utterances['interlocutor'] == 'therapist']['transcript_id'].unique()

# For each such conversation, take utterance_ids 0,1,2,3
therapist_openers = AnnoMI_df[
    (AnnoMI_df['transcript_id'].isin(therapist_first_ids)) &
    (AnnoMI_df['utterance_id'].isin([0, 1, 2, 3]))
].copy()

# Optional: sort for easier reading
therapist_openers = therapist_openers.sort_values(['transcript_id', 'utterance_id']).reset_index(drop=True)

# Preview the result
print(therapist_openers.head(12))
print(f"Number of conversations: {len(therapist_first_ids)}")
print(f"Subset shape: {therapist_openers.shape}")

therapist_openers.to_csv('therapist_openers.csv', index=False)
```
